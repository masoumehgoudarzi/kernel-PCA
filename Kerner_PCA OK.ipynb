```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_moons(n_samples=200, random_state=1234)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y);
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], c='red');
plt.scatter(X[y==1, 0], X[y==1, 1], c='blue');
```

```python
from sklearn.decomposition import PCA , KernelPCA
```

```python
pca_1 = PCA(n_components=1)
pca_2 = PCA(n_components=2)
```

```python
X_1d_pca = pca_1.fit_transform(X)
```

```python
plt.scatter(X_1d_pca[y==0, 0], np.ones(X_1d_pca[y==0, 0].shape)+0.01, c='red', alpha=0.1)
plt.scatter(X_1d_pca[y==1, 0], np.ones(X_1d_pca[y==1, 0].shape)-0.01, c='blue', alpha=0.1)
plt.ylim(0.5, 1.5);
plt.xlabel('PCA 1')
```

```python
kpca_1 = KernelPCA(n_components=1, kernel='rbf', gamma=20, degree=4)
kpca_2 = KernelPCA(n_components=2, kernel='rbf', gamma=20, degree=4)
```

```python
X_1d_kpca = kpca_1.fit_transform(X)
```

```python
plt.scatter(X_1d_kpca[y==0, 0], np.ones(X_1d_kpca[y==0, 0].shape)+0.01, c='red', alpha=0.1)
plt.scatter(X_1d_kpca[y==1, 0], np.ones(X_1d_kpca[y==1, 0].shape)-0.01, c='blue', alpha=0.1)
plt.ylim(0.5, 1.5);
plt.xlabel('Kernel PCA 1');
```

```python
X_2d_kpca = kpca_2.fit_transform(X)
```

```python
plt.scatter(X_2d_kpca[y==0, 0], X_2d_kpca[y==0, 1], c='red', alpha=0.1)
plt.scatter(X_2d_kpca[y==1, 0], X_2d_kpca[y==1, 1], c='blue', alpha=0.1)
```

```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA
import numpy as np
```

```python
X, y = fetch_openml('mnist_784', return_X_y=True)
```

```python
X = X.to_numpy()
```

```python
X = X[:10000, :]
```

```python
y = y[:10000]
```

```python
indx = 100
print(y[indx])
plt.imshow(np.reshape(X[indx, :], (28, 28)));
```

```python
from sklearn.preprocessing import MaxAbsScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
```

```python
# steps = [('scaler', MaxAbsScaler()),
#          ('dimension reduction', KernelPCA(n_components=100, kernel='rbf', gamma=10)),
#          ('classifier', RFC(n_estimators=5))]

steps = [('scaler', MaxAbsScaler()),
         ('dimension reduction', PCA(n_components=0.9)),
         ('classifier', RFC(n_estimators=10))]

clf = Pipeline(steps=steps)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(accuracy_score(y_true=y_test, y_pred=y_pred))
```
